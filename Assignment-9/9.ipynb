```python
import cv2
import numpy as np
from PIL import Image, ImageDraw

# Create frames
frames = []

for i in range(30):
    img = Image.new("RGB", (500, 500), color="white")
    draw = ImageDraw.Draw(img)

    # Moving circle animation
    x = 50 + i*10
    y = 250
    draw.ellipse((x, y, x+50, y+50), fill="blue")

    frames.append(np.array(img))

# Create video
height, width, _ = frames[0].shape
video = cv2.VideoWriter("animation.mp4", cv2.VideoWriter_fourcc(*'mp4v'), 10, (width, height))

for frame in frames:
    video.write(frame)

video.release()

print("Video created successfully!")
```

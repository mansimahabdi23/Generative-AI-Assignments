```python
from transformers import pipeline

# Loading a model fine-tuned on the SQuAD (Stanford Question Answering Dataset)
qa_model = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

print("Model loaded successfully!")
```

```python
def ask_chatbot(context, question):
    result = qa_model(question=question, context=context)
    return result['answer'], round(result['score'], 4)

# --- Example Usage ---
context_text = """
The Great Pyramid of Giza is the oldest and largest of the pyramids in the Giza pyramid complex.
It was built for the Pharaoh Khufu over a 20-year period concluding around 2560 BC.
It stands at approximately 146.6 meters (481 feet).
"""

user_query = "Who was the Great Pyramid built for?"

answer, confidence = ask_chatbot(context_text, user_query)

print(f"Question: {user_query}")
print(f"Answer: {answer}")
print(f"Confidence Score: {confidence}")
```

```python
print("--- Simple QA Chatbot ---")
context = input("Enter the context paragraph: ")

while True:
    question = input("\nAsk a question (or type 'exit' to quit): ")
    if question.lower() == 'exit':
        break

    answer, score = ask_chatbot(context, question)
    print(f"Chatbot: {answer} (Confidence: {score})")
```

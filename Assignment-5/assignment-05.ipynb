# Creative Artwork Generation using Stable Diffusion


## Objective
To generate creative artwork using a pre-trained Stable Diffusion model.

```python
# Install required libraries
!pip install diffusers transformers accelerate torch
```

```python
from diffusers import StableDiffusionPipeline
import torch
```

## Load Stable Diffusion Model

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

## Generate Artwork

```python
prompt = "A futuristic city with neon lights, cyberpunk style, highly detailed"

image = pipe(prompt).images[0]
image.save("generated_art.png")

image
```

## Experiment with Prompts
Try different prompts to generate unique artwork.

```python
prompt2 = "A fantasy dragon flying over mountains at sunset, digital art"
image2 = pipe(prompt2).images[0]
image2
```

Stable Diffusion allows us to generate high-quality creative images from text prompts.
